```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
root_dir = "data"
df = pd.read_csv(f'{root_dir}/processed_asimow_dataset.csv')
df.head()
```

```python
df_columns = df.columns
df_columns
```

```python
df = df[df.labels != -1]
df.head()
```

```python
# df = df.drop(columns=['welding_run'])
```

```python
df["ID"] = np.arange(len(df))
```

```python
df_c = df[['ID', 'labels', 'experiment', 'welding_run'] + [f"I_{i}" for i in range(200)]]
df_c.head()
```

```python
df_v = df[['ID', 'labels', 'experiment', 'welding_run'] + [f"V_{i}" for i in range(200)]]
df_v.head()
```

```python
# melt df with ID, label, and value columns voltage containing V_0, ..., V_199 and current containing I_0, ..., I_199
df_vt = pd.melt(df_v, id_vars=['ID', 'experiment', 'welding_run', 'labels'], var_name='feature', value_name='value')
df_vt.drop(columns=['feature'], inplace=True)
df_vt.rename(columns={'value': 'voltage'}, inplace=True)
df_vt['t_i'] = np.arange(len(df_vt))
df_vt = df_vt.sort_values(by=['ID', 't_i'])
# df_vt.set_index('ID', inplace=True)
df_vt = df_vt[['ID', 'experiment', 'welding_run', 'labels', 'voltage']]
df_vt.reset_index(drop=True, inplace=True)
df_vt.head()
```

```python
# melt df with ID, label, and value columns voltage containing V_0, ..., V_199 and current containing I_0, ..., I_199
df_ct = pd.melt(df_c, id_vars=['ID', 'experiment', 'welding_run', 'labels'], var_name='feature', value_name='value')
df_ct.drop(columns=['feature'], inplace=True)
df_ct.rename(columns={'value': 'current'}, inplace=True)
df_ct['t_i'] = np.arange(len(df_ct))
df_ct = df_ct.sort_values(by=['ID', 't_i'])
# df_ct.set_index('ID', inplace=True)
df_ct = df_ct[['ID', 'experiment', 'welding_run', 'labels', 'current']]
df_ct.reset_index(drop=True, inplace=True)
df_ct.head()
```

```python
plt.plot(df_ct[df_ct.ID == 50]['current'].values)
plt.show()
```

```python
# join both df_vt and df_ct 
len(df_ct), len(df_vt)
```

```python
df_ct['voltage'] = df_vt['voltage']
df_ct
```

```python
def get_val_test_ids():
    return {
        'test_ids': (
            (3, 32),
            (3, 18),
            (1, 27),
            (3, 19),
            (3, 17),
            (2, 21),
            (1, 20),
            (1, 11)
        ),
        'val_ids': (
            (3, 3),
            (2, 10),
            (1, 24),
            (3, 24),
            (1, 32),
            (2, 1),
            (1, 10),
            (1, 16)
        )
    }


val_test_ids = get_val_test_ids()
val_ids = val_test_ids['val_ids']
test_ids = val_test_ids['test_ids']
```

```python
# filter out the validation and test data

val_cond = [(df_ct['experiment'] == exp_id) &(df_ct['welding_run'] == weld_id) for exp_id, weld_id in val_ids]

df_val = df_ct[np.any(val_cond, axis=0)]

test_cond = [(df_ct['experiment'] == exp_id) &(df_ct['welding_run'] == weld_id) for exp_id, weld_id in test_ids]
df_test = df_ct[np.any(test_cond, axis=0)]

df_train = df_ct[~np.any(val_cond, axis=0) & ~np.any(test_cond, axis=0)]
```

```python
df_test[(df_test['experiment'] == 3) & (df_test['welding_run'] == 32)]
```

```python
# save the data
df_train.to_csv(f'{root_dir}/asimow/train.csv', index=False)
df_val.to_csv(f'{root_dir}/asimow/val.csv', index=False)
df_test.to_csv(f'{root_dir}/asimow/test.csv', index=False)
```
